```python
from skimage import io
import datetime as dt
from matplotlib.pyplot import imshow
```

```python
class Pessoa:
    __slots__ = ["_nome", "_cpf", "_addres", "_telefone"]
    def __init__(self, nome,  cpf, endereco, tel):
        self._nome = nome
        self._cpf = cpf
        self._addres = endereco
        self._telefone = tel
##GET
    @property
    def nome(self):
        return self._nome

    @property
    def cpf(self):
        return self._cpf
    @property
    def addres(self):
        return self._addres
    @property
    def telefone(self):
        return self._telefone
##SET
    @nome.setter
    def nome(self, nome):
        self._nome = nome

    @cpf.setter
    def cpf(self, cpf):
        self._cpf = cpf

    @addres.setter
    def addres(self, addres):
        self._addres = addres

    @telefone.setter
    def telefone(self, tel):
        self._telefone = tel

    def imprimir(self):
        print("nome: ", self.nome)
        print("cpf: ", self.cpf)
        print("Endere√ßo: ", self._addres)
        print("Telefone: ", self.telefone)

```

```python
class Fotografia:

    _qtd_fotos = 0
    __slots__ = ["_foto", "_fotografo", "_proprietario", "data"]
    def __init__(self, foto, fotografo, proprietario):
        self._foto = io.imread(foto)
        self._fotografo = fotografo
        self._proprietario = proprietario
        self.data = dt.date.today()
        
        Fotografia._qtd_fotos += 1

    @property
    def foto(self):
        return self._foto
    @property
    def fotografo(self):
        return self._fotografo
    @property
    def proprietario(self):
        return self._proprietario
   

    
    
    @foto.setter
    def foto(self, novafoto):
        self._foto = io.imread(novafoto)

    @fotografo.setter
    def fotografo(self, fotografo):
        self._fotografo = fotografo

    @proprietario.setter
    def proprietario(self, prop):
        self.foto.setter = prop
        
    @staticmethod
    def get_qtd_fotos():
        return Fotografia._qtd_fotos

    def mostrar_fotografia(self):
        imshow(self.foto)

    def prop_foto(self):
        print("Tamanho: ", self.foto.shape)
        self.fotografo.imprimir()
        print("Data: ", str(self.data))
```

```python
pessoa1 = Pessoa("Saul", "123", "Av", "11111111")
pessoa2 = Pessoa("Narciso", "321", "Av", "222222222")
```

```python
fotografia = Fotografia('img.jpg', pessoa1, pessoa2)
```

```python
fotografia.mostrar_fotografia()
```

```python
fotografia.prop_foto()
```
